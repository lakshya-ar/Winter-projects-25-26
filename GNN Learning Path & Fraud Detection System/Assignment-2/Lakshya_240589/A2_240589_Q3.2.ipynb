```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def step(x):
    if (x<0.5):
        return 0
    else:
        return 1
```

```python
x1=np.array([0,0,1,1])
x2=np.array([0,1,0,1])
y=np.array([0,1,1,0])

# Hidden Neuron 1
wh1_1 = np.random.randn()
wh2_1 = np.random.randn()
bh1 = np.random.randn()

# Hidden Neuron 2
wh1_2 = np.random.randn()
wh2_2 = np.random.randn()
bh2 = np.random.randn()

#output neuron
w1=np.random.randn()
w2=np.random.randn()
b=np.random.randn()

lr=0.5
epochs=100000
```

```python
for _ in range (epochs):
    y_cap=np.array([])
    for i in range (len(y)):
        h1 = sigmoid((wh1_1 * x1[i]) + (wh2_1 * x2[i]) + bh1)
        h2 = sigmoid((wh1_2 * x1[i]) + (wh2_2 * x2[i]) + bh2)
        y_cap = np.append(y_cap, sigmoid((w1 * h1) + (w2 * h2) + b))

    gradient_w1=0
    gradient_w2=0
    gradient_b=0
    gradient_wh1_1=0
    gradient_wh1_2=0
    gradient_bh1=0
    gradient_wh2_1=0
    gradient_wh2_2=0
    gradient_bh2=0
    for j in range(len(y)):

        h1 = sigmoid((wh1_1 * x1[j]) + (wh2_1 * x2[j]) + bh1)
        h2 = sigmoid((wh1_2 * x1[j]) + (wh2_2 * x2[j]) + bh2)

        gradient_wh1_1 += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j]) * w1 * h1 * (1 - h1) * x1[j]
        gradient_wh2_1 += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j]) * w1 * h1 * (1 - h1) * x2[j]
        gradient_bh1   += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j]) * w1 * h1 * (1 - h1)

        gradient_wh1_2 += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j]) * w2 * h2 * (1 - h2) * x1[j]
        gradient_wh2_2 += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j]) * w2 * h2 * (1 - h2) * x2[j]
        gradient_bh2   += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j]) * w2 * h2 * (1 - h2)
        
        gradient_w1 += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j]) * h1
        gradient_w2 += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j]) * h2
        gradient_b  += (y_cap[j] - y[j]) * y_cap[j] * (1 - y_cap[j])

    w1 -= (lr * gradient_w1 / len(y))
    w2 -= (lr * gradient_w2 / len(y))
    b  -= (lr * gradient_b / len(y))
    
    wh1_1 -= (lr * gradient_wh1_1 / len(y))
    wh2_1 -= (lr * gradient_wh2_1 / len(y))
    bh1   -= (lr * gradient_bh1 / len(y))
    
    wh1_2 -= (lr * gradient_wh1_2 / len(y))
    wh2_2 -= (lr * gradient_wh2_2 / len(y))
    bh2   -= (lr * gradient_bh2 / len(y))
```

```python
print("Final Predictions:")
for p in range(len(y)):
    h1 = sigmoid((wh1_1 * x1[p]) + (wh2_1 * x2[p]) + bh1)
    h2 = sigmoid((wh1_2 * x1[p]) + (wh2_2 * x2[p]) + bh2)
    final = step(sigmoid((w1 * h1) + (w2 * h2) + b))
    print(f"XOR ({x1[p]} , {x2[p]}) = {final}")
```
