```python
import numpy as np
```

```python
class LinearSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.001, n_iters=2000):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None
    def fit(self, X, y):
        if len(X.shape) == 1:
            X = X.reshape(-1, 1)
            
        n_samples, n_features = X.shape

        y = np.where(y <= 0, -1, 1)

        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.n_iters):
            for i in range(n_samples):
                x_i = X[i]
                y_i = y[i]

                condition = y_i * (np.dot(self.w, x_i) - self.b)

                if condition >= 1:
                    self.w = self.w - self.lr * (2 * self.lambda_param * self.w)
                else:
                    self.w = self.w - self.lr * (
                        2 * self.lambda_param * self.w - y_i * x_i
                    )
                    self.b = self.b - self.lr * (y_i)
    def predict(self, X):
        if len(X.shape) == 1:
            X = X.reshape(-1, 1)
        linear_output = np.dot(X, self.w) - self.b
        return np.sign(linear_output)
```

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([-1, -1, 1, 1, 1]) 

clf = LinearSVM()
clf.fit(X, y)

print("Weights:", clf.w)
print("Bias:", clf.b)
print("Prediction:", clf.predict(X))
```
